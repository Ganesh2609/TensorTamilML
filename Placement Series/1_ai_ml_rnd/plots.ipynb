```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
THETA = 28.155968
M =  0.021652
X = 54.906662
```

```python
data = pd.read_csv('xy_data.csv')
x_data = data['x'].values
y_data = data['y'].values
```

```python
t = np.linspace(6, 60, 1000)

theta_rad = np.deg2rad(THETA)
x_curve = t * np.cos(theta_rad) - np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.sin(theta_rad) + X
y_curve = 42 + t * np.sin(theta_rad) + np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.cos(theta_rad)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(x_data, y_data, c='red', s=10, alpha=0.5, label='Data Points')
plt.plot(x_curve, y_curve, 'b-', linewidth=2, label='Fitted Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Curve Fit: ={THETA}, M={M}, X={X}')
plt.legend()
plt.grid(True)
plt.show()
```
