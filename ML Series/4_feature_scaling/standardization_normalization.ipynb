```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
iris_data = sns.load_dataset('iris')
```

```python
iris_data.head()
```

```python
def min_max_scaler(x):
    mini = min(x)
    maxi = max(x)
    diff = maxi - mini
    scaled = [(i-mini)/(diff) for i in x]
    return scaled
```

```python
iris_data_custom_scaled = iris_data.copy()
```

```python
for i in iris_data_custom_scaled.columns[:-1]:
    iris_data_custom_scaled[i] = min_max_scaler(iris_data_custom_scaled[i])
```

```python
iris_data_builtin_transformed = iris_data.copy().iloc[:, :-1]
iris_data_builtin_transformed.head()
```

```python
x_train, x_test = train_test_split(iris_data_builtin_transformed, train_size=0.8)
```

```python
iris_data_builtin_transformed.iloc[:, 1].min(), x_train.iloc[:, 1].min(), x_test.iloc[:, 1].min()
```

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train_transformed = pd.DataFrame(scaler.transform(x_train), columns=x_train.columns)
x_train_transformed
```

```python
x_test_transformed = pd.DataFrame(scaler.transform(x_test), columns=x_test.columns)
x_test_transformed
```
