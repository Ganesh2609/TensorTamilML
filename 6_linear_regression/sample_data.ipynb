```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
# y = 3x + 5
y = 3*x + 5
y
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=80, color='blue')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
X = np.linspace(0, 10, 1000)
Y = 3*X + 5
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=80, color='blue')
plt.plot(X, Y, color='red')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```
